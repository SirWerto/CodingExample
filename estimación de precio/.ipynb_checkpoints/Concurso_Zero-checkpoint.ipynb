```python
# Imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras import optimizers
from keras.utils import plot_model
import time
from datetime import datetime
from math import pi,sin
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv("train.csv")
data = data.astype({'hour' : 'int32', 'transport_mode' : 'int32', 'price' : 'int32', 'eta' : 'int32'})
```

```python
data['date'] = pd.to_datetime(data['date'], format="%Y-%m-%d")
data['weekday'] = data['date'].dt.date.apply(lambda x: x.weekday())
data['weekend'] = np.where(data['weekday']>=5, 1, 0)
# wT = 2pi -> w = 0.8975979010256552 para hacerla circular
data['weekday'] = data['weekday'].apply(lambda x: sin(x*0.8975979010256552))
#data['month'] = data['date'].dt.date.apply(lambda x: x.month)
del data['date']
del data['sid']
```

```python
#Put bT at the end
cols = data.columns.tolist()
cols.append(cols.pop(cols.index('weekend')))
cols.append(cols.pop(cols.index('transport_mode')))
cols.append(cols.pop(cols.index('binaryTarget')))
data = data[cols]
```

```python
#zeros = data[data['price'] == 0]
zeros = data
ana = data[(data['price'] > 0) & (data['price'] <= 1000)]
nzeros = data[data['price'] > 1000]
```

```python
len(zeros)
```

```python
len(nzeros)
```

```python
datas = zeros[zeros.columns[:-1]]
datas = (datas-datas.mean())/datas.mad()
datas['weekend'] = zeros['weekend']
datas['transport_mode'] = zeros['transport_mode']
datas['BT'] = zeros['binaryTarget']
#del datas['price'], datas['min_price']
ndatas = nzeros[nzeros.columns[:-1]]
ndatas = (ndatas-ndatas.mean())/ndatas.mad()
ndatas['weekend'] = nzeros['weekend']
ndatas['transport_mode'] = nzeros['transport_mode']
ndatas['BT'] = nzeros['binaryTarget']
adatas = ana[ana.columns[:-1]]
adatas = (adatas-adatas.mean())/adatas.mad()
adatas['weekend'] = ana['weekend']
adatas['transport_mode'] = ana['transport_mode']
adatas['BT'] = ana['binaryTarget']
```

```python
prop = 0.3/0.5
temp, test = train_test_split(datas, test_size=0.5)
train, val = train_test_split(temp, test_size=prop)
del temp
```

```python
x_train = train[train.columns[:-1]].values
y_train = to_categorical(train["BT"])
x_test = test[test.columns[:-1]].values
y_test = to_categorical(test["BT"])
validation = (val[val.columns[:-1]].values, to_categorical(val["BT"]))
```

```python
paco = Sequential()
paco.add(Dense(units=20, activation='sigmoid', input_dim=16))
paco.add(Dense(units=15, activation='sigmoid'))
paco.add(Dense(units=10, activation='sigmoid'))
paco.add(Dense(units=15, activation='sigmoid'))
paco.add(Dense(units=2, activation='softmax'))
```

```python
sgd = optimizers.SGD(lr=0.03, clipvalue=0.5, momentum=0.2)
```

```python
paco.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
history = paco.fit(x_train, y_train, epochs=200, batch_size=20, validation_data=validation, verbose=2)
```

```python
paco.evaluate(x_test, y_test, batch_size=15)
```
