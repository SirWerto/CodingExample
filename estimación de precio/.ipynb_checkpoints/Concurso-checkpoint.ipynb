Partimos las muestras en 3 categorías

```python
sample = []
sample.append(data[data['price'] == 0])
sample.append(data[(data['price'] > 0) & (data['price'] <= 1000)])
sample.append(data[data['price'] > 1000])
```

```python
def bt_end(data):
    cols = data.columns.tolist()
    cols.append(cols.pop(cols.index('day')))
    cols.append(cols.pop(cols.index('weekend')))
    cols.append(cols.pop(cols.index('transport_mode')))
    cols.append(cols.pop(cols.index('binaryTarget')))
    data = data[cols]
    return data
```

```python
def split(data):
    prop = 0.2/0.8
    temp, test = train_test_split(data, test_size=0.2)
    train, val = train_test_split(temp, test_size=prop)
    return train, val, test
```

```python
def t_t(train, val, test):
    x_train = train[train.columns[:-1]].values
    y_train = train["binaryTarget"].values
    x_val = val[val.columns[:-1]].values
    y_val = val["binaryTarget"].values
    x_test = test[test.columns[:-1]].values
    y_test = test["binaryTarget"].values
    return x_train, y_train, x_val, y_val, x_test, y_test
```

```python
datas = data[data.columns[:-1]]
datas = (datas-datas.mean())/datas.mad()
datas['weekend'] = data['weekend']
datas['transport_mode'] = data['transport_mode']
datas['binaryTarget'] = data['binaryTarget']
```

```python
nd = []
for d in sample:
    temp = bt_end(d)
    train, val, test = split(temp)
    nd.append(t_t(train, val, test))
```

```python
#Put bT at the end
cols = data.columns.tolist()
cols.append(cols.pop(cols.index('day')))
cols.append(cols.pop(cols.index('weekend')))
cols.append(cols.pop(cols.index('transport_mode')))
cols.append(cols.pop(cols.index('binaryTarget')))
data = data[cols]
prop = 0.2/0.8
temp, test = train_test_split(data, test_size=0.2)
train, val = train_test_split(temp, test_size=prop)
del temp
x_train = train[train.columns[:-1]].values
y_train = train["binaryTarget"].values
x_val = val[val.columns[:-1]].values
y_val = val["binaryTarget"].values
x_test = test[test.columns[:-1]].values
y_test = test["binaryTarget"].values
```

```python
#Put bT at the end
cols = pred.columns.tolist()
cols.append(cols.pop(cols.index('day')))
cols.append(cols.pop(cols.index('weekend')))
cols.append(cols.pop(cols.index('transport_mode')))
pred = pred[cols]
```

Random Forest

```python
accur = 0
cum = 0
for i in range(3):
    
    rf = RandomForestClassifier(n_estimators=150 ,max_depth=None, verbose=0, max_features="sqrt", n_jobs=-1)
    rf.fit(nd[i][0], nd[i][1])
    r_val = rf.score(nd[i][2], nd[i][3])
    r_test = rf.score(nd[i][4], nd[i][5])
    print("validation: " + str(r_val) + " test: " + str(r_test) + " elements " + str(len(nd[i][4])))
    accur = accur + r_test*len(nd[i][4])
    cum = cum + len(nd[i][4])
accur = accur/cum
print("Total accuracity: " + str(accur))
```
