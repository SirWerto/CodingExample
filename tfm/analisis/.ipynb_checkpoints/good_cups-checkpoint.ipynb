```python
#import
import numpy as np
import pandas as pd
import datetime

from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering, KMeans
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet
from scipy.spatial.distance import pdist

import matplotlib
import matplotlib.pyplot as plt

from math import pi

import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
#Variables de decisión
nPCA = 12
nClusters = 11
nSeeds_kMeans = 1000
```

```python
consumoHorasFestivos = pd.read_csv('cups_normales.csv')
consumoHorasFestivos.drop(columns=['clase'], inplace=True)
consumoHorasFestivos.set_index('cups', inplace=True)
consumoHorasFestivos
```

```python
#Componentes principales
pca = PCA(svd_solver='full', n_components=nPCA)
#pca = PCA(svd_solver='full')
X =pca.fit_transform(consumoHorasFestivos)
pca
```

```python
fig, ax = plt.subplots()
ax.plot(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.cumsum())
```

```python
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('número de componentes')
plt.ylabel('Varianza de cada componente')
plt.title('Componentes PCA')
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.cumsum())
```

```python
#Conglomerate cluster and Dendogram
cut_off = 0.65
Z = linkage(X, 'ward', optimal_ordering=True)
fig = plt.figure(figsize=(30,10))
plt.axhline(y=cut_off, c='k')
dn = dendrogram(Z)
```

```python
fig = plt.figure(figsize=(15,5))
plt.axhline(y=cut_off, c='k')
dt = dendrogram(Z, truncate_mode='lastp', p=nClusters+3)
```

```python
print(Z[-nClusters-3:,2])
```

```python
#Cophenet
c, coph_dists = cophenet(Z, pdist(X))
c
```

```python
#K-means clustering
#Init
kmeans = KMeans(n_clusters=nClusters, init='random').fit(X) 

#Multiple seeds
for seed in range(nSeeds_kMeans):
    seed_try = KMeans(n_clusters=nClusters, init='random').fit(X)
    if seed_try.inertia_ < kmeans.inertia_:
        kmeans = seed_try
```

```python
kmeans.inertia_
```

```python
#Applaing groups
media_consumos = consumoHorasFestivos.mean()
consumoHorasFestivos['clase'] = kmeans.labels_
```

```python
Categorias = consumoHorasFestivos.groupby('clase').mean()
Categorias
```

```python
#Preparar plots
horas = list(Categorias.columns)
horas2 = [x[x.find('h')+1:] for x in horas]
values = [Categorias.loc[x].values.flatten().tolist() for x in range(nClusters)]
my_palette = plt.cm.get_cmap("Set2", len(Categorias.index))
```

```python
def make_ploto(row, title, color, horas2, maximo, values, number):
    # number of variable
    categories=horas2
    N = len(categories)

    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
 
    # Initialise the spider plot
    ax = plt.subplot(round(number/3)+1,3, row+1, polar=True)
 
    # If you want the first axis to be on top:
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)
 
    # Draw one axe per variable + add labels labels yet
    plt.xticks(angles[:-1], categories)
 
    # Draw ylabels
    ax.set_rlabel_position(0)
    #plt.yticks([1,2,3], ["1","2","3"], color="grey", size=7)
    plt.ylim(0,maximo)
    
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=1, linestyle='solid')
    ax.fill(angles, values, 'b', color=color, alpha=1)
 
    # Add a title
    plt.title(title, size=23, color=color)
```

```python
plt.figure(figsize=(25,45))
for row in range(0, nClusters):
    make_ploto(row, 'grupo '+str(row), my_palette(row), horas2, max(values[row]), values[row][:], nClusters)
```

```python
#Freq
freq = pd.value_counts(consumoHorasFestivos['clase'])
plt.bar(range(len(freq)), freq.values)
locs, labels = plt.xticks(np.arange(nClusters), freq.index)
plt.xlabel('Grupos')
plt.ylabel('Cups')
plt.title('Cups por grupo')
```
