```python
Recs.isnull().sum()
```

```python
P = pd.read_csv("/home/jorge/tfm-jvm/data/clean/player_features.csv")
P.head()
```

```python
Recs = Recs.merge(P[["id_", "id_p", "id_s"]], how="left", left_on=["id_s", "player"], right_on=["id_s", "id_p"])
Recs["id_p"] = Recs["id_"]
del Recs["id_"], Recs["player"], P
Recs
```

```python
Recs.isnull().sum()
```

```python
Recs.dropna(inplace=True)
```

```python
S = pd.read_csv("/home/jorge/tfm-jvm/data/clean/servers_features.csv")
S.head()
```

```python
Recs = Recs.merge(S[["_id", "id_s", "start_day"]], how="left", left_on=["id_s"], right_on=["_id"])
Recs["id_s"] = Recs["id_s_y"]
del Recs["_id_x"], Recs["id_s_x"], Recs["_id_y"], Recs["id_s_y"], S
Recs
```
