```python
rp = pd.read_csv("/home/jorge/tfm-jvm/data/tables/rec_players.csv")
rp = rp.sort_values(by=["id_p", "rs_day"])
rp
```

```python
def eval_inactive(x, n):
    l = len(x)
    if l < n:
        return pd.Series([False]*l, name="activity_cut")
    i = 0
    recs = []
    flag = False
    while i < l-(n-1):
        if flag:
            recs.append(False)
            i += 1
        else:
            if any(rec == 0 for rec in x[i:i+n]):
                flag = True
                recs.append(False)
                i += 1
            else:
                recs.append(True)
                i += 1
    if flag:
        addr = [False] * (n-1)
    else:
        addr = [True] * (n-1)
    recs.extend(addr)
    if len(recs) != l:
        raise ArithmeticError
    return pd.Series(recs, name="activity_cut")   
```

```python
%%time
N = 4
rp["activity_cut"] = rp.groupby(by=["id_p"])["one_g"].apply(lambda x: eval_inactive(x.rolling(window=N).sum().values.tolist(), n=N)).reset_index()["one_g"]
```
