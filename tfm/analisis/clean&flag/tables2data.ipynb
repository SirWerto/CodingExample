## Get player Results

```python
def get_data_table(x, n):
    if len(x) < n:
        return None
    data = x.iloc[-n:]
    player = {
        "id_p" : data["id_p"].iloc[0],
        "n_days" : n,
        "N_last" : data["N"].iloc[n-1],
        "N_first" : data["N"].iloc[0],
        "N_diff_positive" : data[data["N_diff"] > 0]["N_diff"].sum(),
        "N_diff_negative" : data[data["N_diff"] < 0]["N_diff"].sum(),
        "pop_sum_mean" : data["pop_sum"].mean(),
        "pop_sum_std" : data["pop_sum"].std(),
        "pop_mean_mean" : data["pop_mean"].mean(),
        "pop_mean_std" : data["pop_mean"].std(),
        "norm2_mean_mean" : data["norm2_mean"].mean(),
        "norm2_mean_std" : data["norm2_mean"].std(),
        #"norm2_std_mean" : data["norm2_std"].mean(),
        #"norm2_std_std" : data["norm2_std"].std(),
        "g_sum_mean" : data["g_sum"].mean(),
        "g_sum_std" : data["g_sum"].std(),
        "g_mean_mean" : data["g_mean"].mean(),
        "g_mean_std" : data["g_mean"].std(),
        #"g_std_mean" : data["g_std"].mean(),
        #"g_std_std" : data["g_std"].std(),
        "romanos" : data["romanos"].iloc[n-1],
        "germanos" : data["germano"].iloc[n-1],
        "galos" : data["galos"].iloc[n-1],
        #"natares" : data["natares"].iloc[n-1],
        "egipcios" : data["egipcios"].iloc[n-1],
        "hunos" : data["hunos"].iloc[n-1],
        "romanos_increase" : data["romanos"].iloc[n-1] - data["romanos"].iloc[0],
        "germanos_increase" : data["germano"].iloc[n-1] - data["germano"].iloc[0],
        "galos_increase" : data["galos"].iloc[n-1] - data["galos"].iloc[0],
        "egipcios_increase" : data["egipcios"].iloc[n-1] - data["egipcios"].iloc[0],
        "hunos_increase" : data["hunos"].iloc[n-1] - data["hunos"].iloc[n-1],
        "rs_day_last" : data["rs_day"].iloc[n-1],
        "p_play" : data[data["one_g"] == True]["one_g"].count()/n,
        "p_play_weekend" : data[(data["one_g"] == True) & (data["weekend"] == True)]["one_g"].count()/data[data["weekend"] == True]["one_g"].count(),
        "p_play_week" : data[(data["one_g"] == True) & (data["weekend"] == False)]["one_g"].count()/data[data["weekend"] == False]["one_g"].count()
        
    }
    
    return player
```

```python
%%time
gp = rp_no_natares.groupby("id_p")
n = 7
temp = gp.apply(lambda x: get_data_table(x, n))
```

```python
temp = [x for x in temp.values.tolist() if x != None]
```

```python
Ana = pd.DataFrame(temp)
Ana
```
